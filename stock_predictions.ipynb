```python
import pandas as pd
import numpy as np
from openbb import obb
import plotly as plt
```

```python
def get_data(ticker):
    stock_data = obb.equity.price.historical(symbol=ticker, provider='yfinance', adjusted=True)
    stocks_data = stock_data.to_df()
    return stocks_data
```

```python
ticker = 'GOOGL'
stockdata = get_data(ticker)
```

```python
from scipy.stats import norm
```

```python
log_returns = np.log(stockdata['adj_close'] / stockdata['adj_close'].shift(1))
mean = log_returns.mean()
var = log_returns.var()
drift = mean - (0.5 * var)

stdev = log_returns.std()
days = 50
trials = 10000
Z = norm.ppf(np.random.rand(days, trials)) #days, trials
```

```python
Z
```

```python
stdev
```

```python
daily_returns = np.exp(drift + stdev * Z)

price_list = np.zeros_like(daily_returns)
price_list.shape
```

```python
stockdata.iloc[-1]['close']
```

```python

price_list[0] = stockdata.iloc[-1]['adj_close']
for t in range(1, days):
    price_list[t] = price_list[t-1]*daily_returns[t]
```
